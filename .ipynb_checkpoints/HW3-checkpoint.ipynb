```python
#1 a)
import numpy as np

X = np.array([[3,1,3],[4,3,2],[5,2,1],[6,1,1]]).T
T = np.array([0,1,2,3])

T = np.vstack([T,np.ones(len(T))]).T
m0,c0 = np.linalg.lstsq(T,X[0],rcond=None)[0]
m1,c1 = np.linalg.lstsq(T,X[1],rcond=None)[0]
m2,c2 = np.linalg.lstsq(T,X[2],rcond=None)[0]
b = [c0,c1,c2]
xp1 = np.array([[X[0][3]+m0],[X[1][3]+m1],[X[2][3]+m2]])
A = [[xp1.item(0)/X.item((0,3)),0,0],[0,xp1.item(1)/X.item((1,3)),0],[0,0,xp1.item(2)/X.item((2,3))]]
print(A)
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(projection='3d')

ax.scatter(X[0],X[1],X[2])
ax.scatter(xp1.item(0),xp1.item(1),xp1.item(2),c='r',marker='^')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

x=[c0,c0+4*m0]
y=[c1,c1+4*m1]
z=[c2,c2+4*m2]

ax.plot3D(x,y,z,'red')
plt.show()
```
